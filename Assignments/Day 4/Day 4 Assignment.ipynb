# Question 2:
How to create a series from a numpy array?


```python
import numpy as np
arr=np.arange(0,101,10)
ser=pd.Series(arr)
ser
```

# Question 3:
How to convert the index of a series into a column of a dataframe?


```python
new=pd.DataFrame(ser,columns=['numbers'])
print(new)
new.reset_index()
```
